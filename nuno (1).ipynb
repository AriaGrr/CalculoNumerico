```python

import pandas as pd
import statsmodels.formula.api as sm
df = pd.read_csv('dados_banco.csv', header=(0))
# eliminar o ponto no nome das columas para usar o sm.ols
#substituindo o _ por nada 
df.columns = [ x.replace('_','') for x in df.columns ]
#trocando letras para facilitar escrita 
df.columns = [x.replace('', 'c') for x in df.columns]
#usando saldo_investimento como variavel dependente, enquanro usamos o salario, a idade, o saldo na conta corrente e o saldo na poupana como termos independentes. 
model = sm.ols(formula="Saldoinvestimento ~ Salario+Saldocc+Saldopoupanca + Idade ", data=df)
result = model.fit()
print(result.summary())
#Como o valor p de salario foi o maior entre todos os termos, ele foi retirado e a tabela foi refeita sem ele . 
```
