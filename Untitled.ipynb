```python
#a) usando idade e salário
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.metrics import r2_score
data = pd.read_csv('dados_banco.csv', header=(0))
#data.head(100);
data = data.to_numpy()
y = data[:,5]
x = data[:,3]
X = np.array(x).reshape(-1, 1)
Y = np.array(y)
modelo = LinearRegression().fit(X, Y)
print(modelo.intercept_)
print (modelo.coef_)
y_pred=modelo.predict(X)
R2 = r2_score(y, y_pred)
print('R2:', R2)
n = np.size(x) 
Sx= np.sum(x)
Sy=np.sum(y)
Sxy=np.sum(x*y)
Sxx=np.sum(x*x)   
a1=(n*Sxy-Sx*Sy)/(n*Sxx-Sx**2)   #Calcula o coeficiente a1 da reta
a0=(Sxx*Sy-Sxy*Sx)/(n*Sxx-Sx**2) #Calcula o coeficiente a0 da reta
#print(a1)
#print(a0)
# mostra os dados
plt.scatter(x, y, color = "b", marker = "o", s = 50) 
  # prediz os valores
y_pred = a0 + a1*x 
    # mostra a reta de regressão
plt.plot(x, y_pred, color = "r") 
  
plt.xlabel('x', fontsize = 15) 
plt.ylabel('y', fontsize = 15) 
plt.show(True) 
#correlacao = np.corrcoef(X, Y)
#print(correlacao)
plt.scatter(X, Y)
plt.plot(X, modelo.predict(X), color = 'red')

```

```python

import pandas as pd
import statsmodels.formula.api as sm
df = pd.read_csv('dados_banco.csv', header=(0))
# eliminar o ponto no nome das columas para usar o sm.ols
df.columns = [ x.replace('_','') for x in df.columns ]
df.columns = [x.replace('ç', 'c') for x in df.columns]
model = sm.ols(formula="Saldoinvestimento ~ Salario+Saldocc+Saldopoupanca + Idade ", data=df)
result = model.fit()
print(result.summary())
```
